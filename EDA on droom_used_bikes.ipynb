```python
#13. How does the resale price vary by kilometers driven?
correlation = data[['kms_driven', 'price']].corr()
print(correlation)
```

The correlation coefficient of -0.313955 between kms_driven and price indicates a moderate negative correlation.
This suggests that as the number of kilometers driven increases, the resale price tends to decrease, which is common in the used vehicle market. Higher kilometers typically mean the bike is more used, which can lower its value due to more wear and tear.
The negative correlation means that, in general, bikes with higher mileage have lower resale prices, though this is not a perfect relationship (there could be exceptions based on other factors like brand, condition, etc.).
